```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import colorcet as cc

curve_data = pd.read_csv("./adiv_curve_data.txt",
    sep="\t")

fig, axes = plt.subplots(ncols=2,
    nrows=2,
    figsize=(15, 15))

observed_otus = sns.lineplot(curve_data,
    x="Seqs/Sample",
    y="observed_otus Ave.",
    hue="SampleID",
    ax=axes[0, 0],
    palette=sns.color_palette(cc.glasbey, n_colors=35),
    legend=False)

observed_otus.set(xlabel="Sequences per sample",
    ylabel="Observed OTUs")

chao1 = sns.lineplot(curve_data,
    x="Seqs/Sample",
    y="chao1 Ave.",
    hue="SampleID",
    ax=axes[0, 1],
    palette=sns.color_palette(cc.glasbey, n_colors=35),
    legend=False)

chao1.set(xlabel="Sequences per sample",
    ylabel="Chao1 Index")

shannon = sns.lineplot(curve_data,
    x="Seqs/Sample",
    y="shannon Ave.",
    hue="SampleID",
    ax=axes[1, 0],
    palette=sns.color_palette(cc.glasbey, n_colors=35),
    legend=False)

shannon.set(xlabel="Sequences per sample",
    ylabel="Shannon Index")

pd_whole = sns.lineplot(curve_data,
    x="Seqs/Sample",
    y="goods_coverage Ave.",
    hue="SampleID",
    palette=sns.color_palette(cc.glasbey, n_colors=35),
    ax=axes[1, 1])

pd_whole.set(xlabel="Sequences per sample",
    ylabel="Good's Coverage Index")

sns.move_legend(pd_whole,
    "upper right",
    bbox_to_anchor=(1.2, 1.8))

plt.figtext(0.1, 0.89, 'A',clip_on=False,color='black', weight='bold',size=22)
plt.figtext(0.52, 0.89, 'B',clip_on=False,color='black', weight='bold',size=22)
plt.figtext(0.1, 0.47, 'C',clip_on=False,color='black', weight='bold',size=22)
plt.figtext(0.52, 0.47, 'D',clip_on=False,color='black', weight='bold',size=22)

plt.savefig("../../05_final_figures/suppl_fig1.tiff",
    dpi=600)

plt.show()
plt.clf()
```
